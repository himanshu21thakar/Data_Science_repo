# ðŸ“Š Introduction to Matplotlib and Seaborn

**Matplotlib** is a popular Python library used for data visualization. It provides a flexible way to create static, animated, and interactive plots like line graphs, bar charts, scatter plots, histograms, and more.

**Seaborn** is built on top of matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies complex visualizations and integrates well with pandas DataFrames.

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=10)

# Add title
plt.title('Pie Chart Example')

# Show the plot
plt.show()
# ðŸ“Œ Importing the Matplotlib library for plotting
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot 1
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

# Plot 2
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

# Add common labels
fig.suptitle('Multiple Plots Example')
plt.show()
# ðŸ“Œ Importing the Matplotlib library for plotting
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot with customizations
plt.plot(x, y, marker='o', linestyle='--', color='purple', linewidth=2, markersize=8, label='Custom Line')

# Add grid, labels, and title
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Plot', fontsize=14)
plt.legend()

# Show the plot
plt.show()
# ðŸ“Œ Importing the Matplotlib library for plotting
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')
plt.savefig('my_plot1.png', dpi=300)
# Show the plot
plt.show()
# ðŸ“Œ Importing the Matplotlib library for plotting
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# ðŸ“Œ Importing the Matplotlib library for plotting
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips
```

```python
# ðŸ“Š Defining data points for plotting
sns.scatterplot(x = 'total_bill',y = 'tip',data = tips)
plt.title('Total bill VS tips')
plt.show()
```

```python
# Line plot
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip (Line Plot)")
plt.show()
```

```python
# Bar plot
sns.barplot(x="day", y="total_bill", data=tips,color='lightblue')
plt.title("Total Bill by Day")
plt.grid(axis = 'y',alpha = 0.3)
plt.show()
```

```python
tips['day'].unique()
```

```python
# Histogram
sns.histplot(tips["total_bill"], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day (Box Plot)")
plt.grid()
plt.show()
```

```python
# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day (Violin Plot)")
plt.show()
```

```python
# Pair plot
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
# ðŸ”¸ Seaborn scatter plot with labeled axes and title
# Facet grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
plt.xlabel("Total Bill")  # X-axis label
plt.ylabel("Tip")  # Y-axis label
plt.title("Scatterplot of Total Bill vs Tip")  # Title for the chart
```

```python
# ðŸ”¸ Seaborn scatter plot with labeled axes and title
# Customizing a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", style="smoker", size="size")
plt.title("Customized Scatter Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Time", loc="upper right")
plt.show()
```

```python
# ðŸ”¸ Seaborn scatter plot with labeled axes and title
# Save a plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.savefig("scatter_plot.png")
plt.xlabel("Total Bill")  # X-axis label
plt.ylabel("Tip")  # Y-axis label
plt.title("Scatterplot of Total Bill vs Tip")  # Title for the chart
```
