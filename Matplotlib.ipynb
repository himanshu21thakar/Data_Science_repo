üîß Getting Started
Before you can use Matplotlib, you need to import it. The most common convention is:

#### import matplotlib.pyplot as plt




---

### 1. **Basic Line Plot**
```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Prime Numbers')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()

# Show the plot
plt.show()
```

---

### 2. **Bar Chart**
```python
import matplotlib.pyplot as plt
```

### 5. **Pie Chart**
```python

colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

```

---

### 6. **Multiple Plots (Subplots)**
```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot 1
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

# Plot 2
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

# Add common labels
fig.suptitle('Multiple Plots Example')
plt.show()
```

---

### 7. **Customizing Plots**
```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot with customizations
plt.plot(x, y, marker='o', linestyle='--', color='purple', linewidth=2, markersize=8, label='Custom Line')

# Add grid, labels, and title
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Plot', fontsize=14)
plt.legend()

# Show the plot
plt.show()
```

---

### 8. **3D Plot**
```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

---

### 9. **Saving a Plot to a File**
```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Saving a Plot')

# Save the plot to a file
plt.savefig('my_plot.png', dpi=300)

# Show the plot
plt.show()
```
